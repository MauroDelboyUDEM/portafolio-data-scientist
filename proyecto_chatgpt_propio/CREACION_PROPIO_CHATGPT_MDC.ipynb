```python
# ---------------------------------------------
# Paso 2: Importar librerías
# ---------------------------------------------
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

```

```python
# ---------------------------------------------
# Paso 3: Definir el modelo y cargarlo
# Modelo: OpenChat 3.5 1210 - No requiere login
# ---------------------------------------------
model_id = "openchat/openchat-3.5-1210"

# Cargar tokenizer y modelo
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",  # usa GPU si está disponible
    torch_dtype=torch.float16,  # más rápido y eficiente
)

```

```python
# ---------------------------------------------
# Paso 4: Crear una función para responder
# ---------------------------------------------
def responder(prompt, max_tokens=200):
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(model.device)
    output = model.generate(
        input_ids,
        max_new_tokens=max_tokens,
        do_sample=True,
        top_p=0.95,
        temperature=0.7,
    )
    respuesta = tokenizer.decode(output[0], skip_special_tokens=True)
    return respuesta

```

```python
# ---------------------------------------------
# Paso 5: Probar con un ejemplo
# ---------------------------------------------
pregunta = """<|system|>Responde con emojis.<|end|>
<|user|>Hola, soy Mauro. Quiero entrar al mundo de la inteligencia artificial. ¿Qué me recomiendas?<|end|>
<|assistant|>"""

respuesta = responder(pregunta)
print(respuesta)
```

```python
# ---------------------------------------------
# Paso 6: Probar con otro ejemplo
# ---------------------------------------------
pregunta = """<|system|>Eres un asesor creativo que usa emojis y lenguaje sencillo.<|end|>
<|user|>Dame 5 ideas para enseñar algoritmos a niños de primaria usando juegos.<|end|>
<|assistant|>"""

respuesta = responder(pregunta)
print(respuesta)
```
